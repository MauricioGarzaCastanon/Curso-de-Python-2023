Tomar el archivo quijote.txt e identificar la cantidad de palabras, caracteres, oraciones y la frecuencia de palabras

```python
archivo ='/content/quijote.txt'
with open(archivo,'r') as f:
  mensaje = f.read()
  lines= f.readlines()
  palabras = mensaje.split(' ')
  oraciones = mensaje.split('.')
  quitar_signos_puntuacion = ",.:;\n!\"'"
  for caracter in quitar_signos_puntuacion:
    mensaje_sin_signos = mensaje.replace(caracter, "")
  palabras_minusculas = mensaje_sin_signos.lower()
  preparar_palabras = palabras_minusculas.split(' ')
  diccionario_auxiliar = {}
  for word in preparar_palabras:
    if word in diccionario_auxiliar:
      diccionario_auxiliar[word] += 1
    else:
      diccionario_auxiliar[word] = 1
  for word in diccionario_auxiliar:
    frecuencia_palabra = diccionario_auxiliar[word]
    print("la palabra ",{word},"aparece",{frecuencia_palabra},"veces en Quijote.txt")
  print('Hay {0} palabras en el archivo Quijote.'.format(len(preparar_palabras)))
  print('Hay {0} letras en el archivo Quijote.'.format(len(mensaje)))
  print('Hay {0} oraciones en el archivo Quijote.'.format(len(oraciones)))
```
